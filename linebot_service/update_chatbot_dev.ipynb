```python
from google.cloud import storage
import os
import magic
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = os.path.abspath('../utility-encoder-420001-0db7ee074ec6.json')
class GoogleCloudStorage:
    def __init__(self, bucket_name):
        self.bucket_name = bucket_name
        self.storage_client = storage.Client()
        self.bucket = self.storage_client.bucket(self.bucket_name)

    def upload_file(self, source_file, destination_blob_name, folder_name=None):
        if folder_name:
            destination_blob_name = folder_name + '/' + destination_blob_name
        blob = self.bucket.blob(destination_blob_name)
        blob.upload_from_file(source_file)

    def delete_file(self, blob_name, folder_name=None):
        if folder_name:
            blob_name = folder_name + '/' + blob_name
        blob = self.bucket.blob(blob_name)
        if blob.exists():
            blob.delete()
        
    def download_file(self, source_blob_name):
        bucket = self.bucket
        blob = bucket.blob(source_blob_name)

        # Fetch the content type
        content_type = blob.content_type
        
        # Download the contents of the blob as bytes
        file_data = blob.download_as_bytes()

        # Determine the content type if it is not set
        if content_type is None:
            mime = magic.Magic(mime=True)
            content_type = mime.from_buffer(file_data)
        return content_type, file_data

content_type, file = GoogleCloudStorage("laoshifu").download_file("1/110版初任主管班實施計畫-1.pdf")
```

```python
import pandas as pd
import io

def read_excel_from_bytes(bytes_data):
    data = io.BytesIO(bytes_data)
    dfs = pd.read_excel(data, sheet_name=None)
    return dfs

def process_for_actom_sheet(df):
    df.columns = df.iloc[0]
    df = df.iloc[1:, 2:4]
    df = df.dropna()
    return df

def process_multipule_sheet(dfs):
    qa = ""
    for sheet_name, df in dfs.items():
        df = process_for_actom_sheet(df)
        qa += (str(df.to_dict(orient='records')))
    return qa
dfs = read_excel_from_bytes(file)
qa = process_multipule_sheet(dfs)
```
